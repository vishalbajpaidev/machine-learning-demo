```python
import numpy as np
import scipy
from matplotlib.pyplot import *
```

```python
x = np.array([0,1,2,3,4,5])
y = np.array([0,0.8,0.9,0.1,-0.8,-1])
```

```python
p1=scipy.polyfit(x,y,1)
p2=scipy.polyfit(x,y,2)
p3=scipy.polyfit(x,y,3)
xp = x
xp=scipy.linspace(-2,6,100)
%matplotlib inline
plot(x,y,'o')
plot(xp,scipy.polyval(p1,xp),'r-')
plot(xp,scipy.polyval(p2,xp),'b--')
plot(xp,scipy.polyval(p3,xp),'y:')
show()
# scipy.polyval?
```
